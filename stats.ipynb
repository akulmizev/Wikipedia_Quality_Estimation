```python
import pandas as pd
import matplotlib as plt
```

```python
article_length = pd.read_csv(f'stats/as/article_length.csv')              #assamese
unique_word_counts = pd.read_csv(f'stats/as/unique_word_counts.csv')
unique_subword_counts = pd.read_csv(f'stats/as/unique_subword_counts.csv')
mean_word_length = pd.read_csv(f'stats/as/mean_word_length.csv')
total_trigrams = pd.read_csv(f'stats/as/total_trigrams.csv')
english_chars = pd.read_csv(f'stats/as/english_chars.csv')
df_as = article_length.merge(unique_word_counts, on='id')
df_as = df_as.merge(unique_subword_counts, on='id')
df_as = df_as.merge(mean_word_length, on='id')
df_as = df_as.merge(total_trigrams, on='id')
df_as = df_as.merge(english_chars, on='id')
df_as
```
